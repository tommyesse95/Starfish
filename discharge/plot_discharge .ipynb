```python
from matplotlib import pyplot as plt 
import pandas as pd
from scipy import interpolate 

# Import data taken with graph reader 


# data_sparc = pd.read_csv('data_sparc.csv', header=None, names=('Pd', 'V')) # Ref: 
data_exp_34 = pd.read_csv('./data_ref/data_exp_34.csv', header=None, names=('Pd', 'V')) # Ref: 
data_analytic = pd.read_csv('./data_ref/data_analytic.csv', header=None, names=('Pd', 'V')) # M = 10 ^ 4 
data_sim = pd.read_csv('./data_ref/data_sim.csv', header=None, names=('Pd', 'V')) 
plt.figure(figsize=(8,5), dpi=100)


# Data from simulations
pd_sparc = [0.8, 1.2, 1.5, 5.5] # Torr cm 
V_break_sparc = [240, 230, 300, 510] # V



# Interpolate data
# f = interpolate.interp1d(pd_sparc, V_break_sparc, kind = 'cubic')


# Plot 
plt.plot (pd_sparc, V_break_sparc, 'ko--', label='SPARC')
plt.plot (data_exp_34.Pd, data_exp_34.V, 'ro--', label='Exp')
plt.plot (data_analytic.Pd, data_analytic.V, 'g*--', label='Ana')
plt.plot (data_sim.Pd, data_sim.V, 'b*--', label='Sim')


# Plot Edit 
plt.xlabel('Pd (Torr cm)')
plt.ylabel('Voltage (V)')
plt.title('Breakdown Voltage for Kr', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.grid(True)


# X, Y axis Tickmarks (scale of your graph) put here the values of Sparc. 
plt.xticks([0,1.5,2.5,3.5,4.5])
plt.yticks([200,260,300,350,450,600])

plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('breakdown_voltage.png', dpi=300)

plt.show()
```
